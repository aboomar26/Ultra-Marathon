#### Does season affect the average speed of players?

```python
# season is the summer slower than winter
# sprin 3-5
# summer 6-8
# fall 9 -11
# winter 12-2
```

```python
df2["month"]=df2["Event dates"].str.split(".").str.get(1).astype(int)
```

```python
df2["race_season"]=df2["month"].apply(lambda x :"winter" if x>11 else "fall" if x>8 else "summer" if x>5 else "sprin" )
```

```python
df2["race_season"]
```

```python
df2.groupby("race_season")['Athlete average speed'].agg(["mean",'count']).sort_values("mean",ascending=False)
```
